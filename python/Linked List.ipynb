```python
def getEfficiency(skills):

  if len(skills) % 2:
    return -1

  sumL = sum(skills)
  sumP = 0
  if sumL%(len(skills)/2):
    return -1
  else:
    sumP = sumL//(len(skills)//2)

  sumP = sumL/(len(skills)/2)
  print(sumL%(len(skills)/2),sumL)
  visited = []
  res = 0
  for i in skills:
    if i in visited:
      continue
    key = sumP - i
    if key in skills:
      res+= (key * i)
      visited.append(i)
      visited.append(key)
  return res

# o(n) o(n)


# skills = [3,2,1,2]
skills =[2,1,1,4,3,5]
getEfficiency(skills)
```

```python
# https://leetcode.com/problems/remove-nth-node-from-end-of-list/
# 19. Remove Nth Node From End of List
# MEDIUM
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
      # add a dummy node with next pointer as head
        dummy = ListNode(0, head)
        left = dummy
        # set right to head
        right = head
        # move to right n times
        while n > 0 and right:
            right = right.next
            n-=1
          # move both left and right pointer until the end of the linkedList
        while right:
            left = left.next
            right = right.next
        # delete means skipping the node
        left.next = left.next.next
        return dummy.next

        # o(n) o(1)

head = [1,2,3,4,5]
n = 2
removeNthFromEnd(head,n)
```

```python
# https://leetcode.com/problems/find-the-duplicate-number/
# 287. Find the Duplicate Number

# Given n array of integers nums containing n + 1 integers where each integer is in the range [1, n] inclusive.
# use FLoyds tortoise and hare algorithm
# basically the ditance to the start of the lop is same for:
# 1. from the start of the array
#2. from the intersection of the two pointers( fast and slow)


def findDuplicate(nums) -> int:
        slow, fast = 0, 0
        # move fast and slow
        while True:
            slow = nums[slow]
            fast = nums[nums[fast]]
            # break on intersection
            if slow == fast:
                break
        # initialzie other slow pointer for the start of the array
        slow2 = 0
        # return on intersection
        while True:
            slow = nums[slow]
            slow2 = nums[slow2]
            if slow == slow2:
                return slow
# o(n) o(1)

nums = [1,3,4,2,2]

findDuplicate(nums)
```

```python
# https://leetcode.com/problems/lru-cache/
# 146. LRU Cache


# node class # doubly linked list node
class Node:
    def __init__(self, key, val):
        self. key, self.val = key, val
        self.prev = self.next = None

class LRUCache:
    # store capacity and chache for storing nodes
    def __init__(self, capacity: int):
        self.cap = capacity
        self.cache = {} # map key to node

        # left = LRU, right = most recent
        # initially put right next to left
        # we keep adding the elements to the right as it will be recently accessed node
        # we remove node next to left as it will be least accessed node
        self.left, self.right = Node(0,0), Node(0,0)
        self.left.next, self.right.prev = self.right, self.left
    # logic for reving node in between two nodes
    def remove(self, node):
        next, prev = node.next, node.prev
        prev.next = next
        next.prev = prev

       # insert alsways to the right as it is recently accessed node
    # logic for adding node to the right
    def insert(self, node):
        prev, next = self.right.prev, self.right
        prev.next = next.prev = node
        node.next, node.prev = next, prev



    def get(self, key: int) -> int:
      # if already in cache
        if key in self.cache:
          # remove from the list
            self.remove(self.cache[key])
            # and inseet to most recently accessed node that is right
            self.insert(self.cache[key])
            # return node value
            return self.cache[key].val
          # else -1
        return -1

    def put(self, key: int, value: int) -> None:
      # if key in cache
        if key in self.cache:
            self.remove(self.cache[key])
            # remove from the linked list
        # add the node to the cache
        self.cache[key] = Node(key, value)
        # put it to the most recently used side that is right side
        self.insert(self.cache[key])
          # check for capacity
          # if full, remove the leftmost node as that is least recently used node
        if len(self.cache)> self.cap:
            lru = self.left.next
            self.remove(lru)
            del self.cache[lru.key]



# push and get at o(1) o(n)

# Your LRUCache object will be instantiated and called as such:
# obj = LRUCache(capacity)
# param_1 = obj.get(key)
# obj.put(key,value)

lru = LRUCache(2)
lru.put(1,1)
lru.put(2,2)
print(lru.get(1))
lru.put(3,3)
print(lru.get(2))
lru.put(4,4)
print(lru.get(1))
print(lru.get(3))
print(lru.get(4))
print(lru.cache)

# # Input
# ["LRUCache", "put", "put", "get", "put", "get", "put", "get", "get", "get"]
# [[2], [1, 1], [2, 2], [1], [3, 3], [2], [4, 4], [1], [3], [4]]


```

```python

# https://leetcode.com/problems/merge-k-sorted-lists/description/
# 23. Merge k Sorted Lists

# HARD
# use merge and sort
# handle two list at an time and sort them


# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
      # handling the edge case where list is empty
        if not lists or len(lists) == 0:
            return None
        # breaking condition is list should have atleast two lists
        while len(lists)>1:
            listRes = []
            # iterate throught, considering two lists at a time
            for i in range(0, len(lists), 2):
                l1 = lists[i]
                # else condition to handle if there is noe other list, then return NOne
                # None and a list can be sorted returning a list
                l2 = lists[i+1] if (i+1) < len(lists) else None
                # add to list Res
                listRes.append(self.mergeTwoLists(l1,l2))
            # replace the lists with listRes
            # lists wil have result sorted linked lists
            lists= listRes
        # return top list
        return lists[0]


# revering the linked list logic
    def mergeTwoLists(self, l1,l2):
        dummy = ListNode()
        tail = dummy
        while l1 and l2:
            if l1.val < l2.val:
                tail.next = l1
                l1 = l1.next
            else:
                tail.next = l2
                l2 = l2.next
            tail = tail.next

        if l1:
            tail.next = l1
        else:
            tail.next = l2
        return dummy.next



# o(nlogk) o(n)
```

```python
# https://leetcode.com/problems/reverse-nodes-in-k-group/
# 25. Reverse Nodes in k-Group

# HARD

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseKGroup(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
      # creating a dummy node with next as head
        dummy = ListNode(0, head)
        groupPrev = dummy

        # add breaking condition
        while True:
          # get the kth node after moving k places
          # it is possible that we return None as kth node, thich means there are no further greoups of k
            kth = self.getKth(groupPrev, k)
            # so we break here as there is no need to reverse if None is returned
            if not kth:
                break
            # store the start of next group
            groupNext = kth.next

            # set curr to groupPrev.next
            # for normal reversing the linked list, we set prev to None
            # but if we so that here, we end up splitting the linked lists
            # None>1>2 reversed will be 2>1>None
            # so here set the prev to kth.next
            # regular reversing logic
            prev, curr = kth.next, groupPrev.next
            while curr != groupNext:
                tmp = curr.next
                curr.next = prev
                prev = curr
                curr = tmp

            # now our kth will be at the start and groupPrev will be at kth
            # store the node at groupPrev.next
            temp = groupPrev.next
            # our groupPrev.next will be the kTh node
            groupPrev.next = kth
            # groupPrev will be the stored node to start the next cycle
            groupPrev = temp

        return dummy.next


# get the kth node
    def getKth(self, curr, k):
        while curr and k>0:
            curr = curr.next
            k-=1
        return curr


# o(n) o(1)
```
