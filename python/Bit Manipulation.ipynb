```python
# https://leetcode.com/problems/counting-bits/
# 338. Counting Bits


# image above
def countBits(n):
        # using dp
        # so get the position of most signinficant bit
        # subtract if from current n and add 1
        # how to change the most significant bit, it is when you know
        # take previous power of 2 and multiply by two
        # if it is equal to n, then update the new power of 2 and use that here onwards
        dp = [0] * (n + 1)
        # highest power of 2 is initially 1
        offset = 1
# go through the range of 1 to n
        for i in range(1, n + 1):
            # we know that we have reached new power of two, update it
            if offset * 2 == i:
                offset = i
            # we add 1 to base problem
            dp[i] = 1 + dp[i - offset]
        return dp



n = 2
print(countBits(n))
```

```python
# https://leetcode.com/problems/reverse-bits/
# 190. Reverse Bits

def reverseBits(n):
        # to store the result
        res = 0
        for i in range(32):
            # start from 0 to 31
            # get the last bit first
            # n& 1 gives n
            # but before to get the ith bit, rotate the n i times
            bit = (n >> i) & 1
            # now move the bit to the ith position bny subtracting it from 31
            # or it to the resul to just get the reverse of the result
            # as n | (0 or) is always n
            res += (bit << (31 - i))
        return res

n = 00000010100101000001111010011100

reverseBits(n)
```

```python
# https://leetcode.com/problems/sum-of-two-integers/
# 371. Sum of Two Integers

# xor makes the different input as 1 or it is zero for the same
# we can have. acarry if 1 + 1

# so first xor a and b
# then to add the carry, we AND a and b and then left shift to move the carry to correct position
# we do this untill we have no carry
#o(n) o(1)

def getSum(a,b):
        # this means we are using b as carry
        while b:
            # we calculate for b first as we will be changing a and we will be needing
            # a for b calculation
            temp = (a & b) <<1
            a = a^b
            b = temp
        return a


a = 1
b = 2
getSum(a,b)
```
