```python
# https://leetcode.com/problems/merge-intervals/description/
# 56. Merge Intervals

#56. MEDIUM



def merge(intervals):
        # sort according to the first number
        intervals.sort(key=lambda pair: pair[0])
        # add first to the res
        output = [intervals[0]]

        for start, end in intervals[1 : ]:
            # get the last element of the output second digit
            lastEnd = output[-1][1]
    # if current start is lesst the last end
            if start <= lastEnd:
                # merge
                output[-1][1] = max(lastEnd, end)
            else:
                # append toe res
                output.append([start, end])
        # return output
        return output
# o(nLogn)




intervals = [[1,3],[2,6],[8,10],[15,18]]
merge(intervals)
```

```python
# https://leetcode.com/problems/non-overlapping-intervals/
# 435. Non-overlapping Intervals


def eraseOverlapIntervals(intervals):
        # sort the intervals
        intervals.sort()

        res = 0

        prevEnd = intervals[0][1]
        # remaining list of intervals
        for start, end in intervals[1:]:
            # they are not overlapping
            if start >= prevEnd:
                # will update the last end short current end
                prevEnd = end
            else:
              # we will remove one intrval, the one with longest end time is removed
                res += 1
                # will keep the last end short to avoid overlapping
                prevEnd = min(end, prevEnd)
        return res


#o(nlogn) o(n)



intervals = [[1,2],[2,3],[3,4],[1,3]]
eraseOverlapIntervals(intervals)
```

```python
# https://leetcode.com/problems/minimum-interval-to-include-each-query/
# 1851. Minimum Interval to Include Each Query
# HARD

import heapq

def minInterval(intervals, queries):
        # sort intervals
        intervals.sort()
        # initialize minHeap
        minHeap = []
        # dictionary to store results
        res = {}
        # initialize i
        # we keep incrementing i and not decrementing it
        i = 0
        # iterate throug q
        for q in sorted(queries):
            # if i is less than interval length and qis less than end of ith interval
            while i < len(intervals) and intervals[i][0] <= q:
                l, r = intervals[i]
                # add the difference and right value to the minHeap
                heapq.heappush(minHeap, (r - l + 1, r))
                # move to next interval
                i += 1
                # we do not decrement i as coming queries can have the same i and q is sorted
            # if minHeap and q is greater than the min right
            while minHeap and minHeap[0][1] < q:
                # pop the min
                heapq.heappop(minHeap)
            # add to dictionary is minHeap orelse -1
            res[q] = minHeap[0][0] if minHeap else -1
        # return values in dict as list
        return [res[q] for q in queries]



intervals = [[1,4],[2,4],[3,6],[4,4]]
queries = [2,3,4,5]
minInterval(intervals, queries)
```
