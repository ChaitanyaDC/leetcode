``````````python
# https://leetcode.com/problems/combination-sum/
# 39. Combination Sum
# Medium
def combinationSum(candidates, target):
        # so basically we have decision to add that. number or not add that number
        # base case is target is equal to the sum

        # result
        res = []
        # to add current result
        cur = []
        def dfs(i, total):
          # base case
            if total == target:
                res.append(cur.copy())
                return
              # if total is greater than target
            if i >= len(candidates) or total> target:
                return
              # append to the string
            cur.append(candidates[i])
            # perform backtracking adding the i to total
            dfs(i,total+candidates[i])
            # per forming without adding to the total and not adding the i(that is adding next element)
            cur.pop()
            dfs(i+1, total)

        dfs(0, 0)
        return res
        # O(2**t) where t is the target or height of the tree




candidates = [2,3,5]
target = 8
combinationSum(candidates, target)
``````````

``````````python
# https://leetcode.com/problems/permutations/description/

# 46. Permutations


def permute(nums):
        result = []
        # base case
        if(len(nums) == 1):
            return [nums.copy()] # or nums[:]
        # go through each element in nums one at a time
        for i in range(len(nums)):
            # remove that num
            n = nums.pop(0)
            # get the combinations removing that sum
            perms = permute(nums)
            # add to the resul again that popped n
            for perm in perms:
                perm.append(n)
            # extend the result
            result.extend(perms)
            # add that number again to the end
            # continue popping another digit
            nums.append(n)

        return result

# o(n!) o(n!)
nums = [1,2,3]
permute(nums)

``````````

``````````python
# https://leetcode.com/problems/subsets-ii/
# 90. Subsets II

#Medium

def subsetsWithDup(nums):
        res = []

        nums.sort()

        def backtrack(i, subset):
            if i == len(nums):
                res.append(subset[::])
                return

            # All subsets that include nums[i]
            subset.append(nums[i])
            backtrack(i+1, subset)
            subset.pop()

            # All the subsets that don't include nums[i]
            # we skip the duplicates in this part
            while i+1 < len(nums) and nums[i] == nums[i+1]:
                i+=1
            backtrack(i+1, subset)
        backtrack(0, [])
        return res


nums = [1,2,2]

# nums = [0]


subsetsWithDup(nums)
# o(n 2**n)
``````````

``````````python
# https://leetcode.com/problems/word-search/
# 79. Word Search


def exist(board, word):
  rows, cols = len(board), len(board[0])
  path = set()

  def dfs(r,c,i):
      if i == len(word):
        return True
      if(r<0 or c<0 or r>=rows or c>=cols
        or word[i] != board[r][c] or (r,c) in path):
        return False
      path.add((r,c))
      res = (dfs(r+1,c,i+1) or
            dfs(r-1,c,i+1) or
            dfs(r,c+1,i+1) or
            dfs(r,c-1,i+1))
      path.remove((r,c))
      return res

  for r in range(rows):
    for c in range(cols):
      if dfs(r,c,0):
        return True
    return False

# o(n*m*  4**length of word)

# board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]]
# word = "ABCCED"

board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]]
word = "SEE"
exist(board, word)
``````````
