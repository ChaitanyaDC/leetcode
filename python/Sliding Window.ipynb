```python
# https://leetcode.com/problems/minimum-window-substring/
# 76. Minimum Window Substring
# maintain have and need counts
#
#
#
#
#

def minWindow( s, t):
  if t== "": return ""

  countT, window = {},{}
  for c in t:
    countT[c] = countT.get(c,0)+1
  have, need = 0, len(countT)
  res, resLen = [-1, -1], float("infinity")

  l = 0
  for r in range(len(s)):
    c = s[r]
    window[c] = 1+ window.get(c,0)
    if c in countT and window[c] == countT[c]:
      have+=1
    # while loop as we need to keep on checking, we are basically adding
    #the characters that are not in target string

    while have == need:
      # update result
      if( r-l+1)< resLen:
        res = [l,r]
        resLen = (r-l+1)
      # pop from left side as we need to get the min
      # print(s[l])
      # print(window)
      window[s[l]] -=1
      if s[l] in countT and window[s[l]]< countT[s[l]]:
        have-=1
      l+=1

  l,r = res
  return s[l:r+1] if resLen != float("infinity") else ""



s = "ADOBECODEBANC"
t = "ABC"

# s = "a"
# t = "a"
minWindow(s,t)
```
