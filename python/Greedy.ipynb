```python
# https://leetcode.com/problems/jump-game/
# 55. Jump Game


def canJump(nums):
        # goal is to reach the last index
        goal = len(nums) - 1

    # iterating from secod last i index to o in -1
        for i in range(len(nums) - 2, -1, -1):
            # if the num at index i has min jum count to reach goal
            if i + nums[i] >= goal:
                #  out next goal is to reach this index
                goal = i
            # check if we are at starting index
        return goal == 0
# o(n) o(1)

nums = [2,3,1,1,4]
canJump(nums)
```

```python
# https://leetcode.com/problems/jump-game-ii/
# 45. Jump Game II


def jump(nums):
        # initialize left and right pointer
        l, r = 0, 0
        res = 0
        #
        while r < (len(nums) - 1):
            # how farthest can we get for that portion
            farthest = 0
            # for all the nums in that region, gewt how farthest can we get
            for i in range(l, r + 1):
                farthest = max(farthest, i + nums[i])
            # then move the left to r plus 1 that is start of next group
            l = r + 1
            # right to farthest we can geo from that group
            r = farthest
            # increment the res for each gorup
            res += 1
        return res



nums = [2,3,1,1,4]
jump(nums)


```

```python
# https://leetcode.com/problems/gas-station/
# 134. Gas Station


def canCompleteCircuit(gas, cost):
        total = 0
        res = 0
        # iterate trhought eh gas elngth
        for i in range(len(gas)):
            # get the diff at i and add to the toal
            total+=(gas[i] - cost[i])
            # if the total drops below 0, make the total zero
            # taht means we cannot reach the loop end with that as start position
            if total <0:
                total = 0
                # so set the start as next of i for now
                res = i+1
        return res

#o(n) o(1)

gas = [1,2,3,4,5]
cost = [3,4,5,1,2]

canCompleteCircuit(gas, cost)

```

```python
# https://leetcode.com/problems/hand-of-straights/
# 846. Hand of Straights


import heapq
def isNStraightHand( hand, groupSize) :
        # check i fdivisible
        if len(hand) % groupSize:
            return False
        # create a dict of count of the chars
        count = {}
        for n in hand:
            count[n] = 1 + count.get(n, 0)
        # create a minHeap of keys
        # minheap will always nbe having min value at the start which can be used to create groups
        minH = list(count.keys())
        heapq.heapify(minH)
        # we empty the minHeap at the end
        while minH:
            # get the first minHeap value
            first = minH[0]
            # we start from i that is min value and get the consecutive numbers untill the group size
            for i in range(first, first + groupSize):
                # if i is not in count that means consecutive number is not in dict
                if i not in count:
                    return False
                # decrement the count of i
                count[i] -= 1
                # if count oif i is zero
                if count[i] == 0:
                    # check if the min of the heap is i if count is zero
                    if i != minH[0]:
                        # we can only have the min on the heap to move to count zero
                        # if we have some other elemnt than the min of the heap, then we are
                        # creating a hole
                        return False
                    # pop the min
                    heapq.heappop(minH)
        return True



# hand = [1,2,3,6,2,3,4,7,8]
# groupSize = 3
hand = [1,2,3,4,5]
groupSize = 4
isNStraightHand( hand, groupSize)


# https://leetcode.com/problems/divide-array-in-sets-of-k-consecutive-numbers/
```

```python
# https://leetcode.com/problems/merge-triplets-to-form-target-triplet/
# 1899. Merge Triplets to Form Target Triplet


def mergeTriplets(triplets, target):
        # set to store 3 target numbers
        good = set()
        # iterate throught he triplest list
        for t in triplets:
            # if anyof the triplet value is greater than corresponding positionof the target, then do nothing
            if t[0] > target[0] or t[1] > target[1] or t[2] > target[2]:
                continue
            # else enumerate like 0:2, 1:5, 2:3
            for i, v in enumerate(t):
                # if the value is equal to target value at that index i
                if v == target[i]:
                    # we thaen add to good set
                    good.add(i)
                # we do thius for aevery triplet as we know that we will reach to the solution at some point

        return len(good) == 3
# o(n) o(1)

triplets = [[2,5,3],[1,8,4],[1,7,5]]
target = [2,7,5]

mergeTriplets(triplets, target)
```

```python
# https://leetcode.com/problems/valid-parenthesis-string/
# 678. Valid Parenthesis String

def checkValidString(s):
        # basically we find the max if we were to have left or right parathesis for wildcard
        # expty string for wild card doesn't matter
        leftMin, leftMax = 0, 0
#
        for c in s:
            # if left paranthesis
            if c == "(":
                # increment the count
                leftMin, leftMax = leftMin + 1, leftMax + 1
            elif c == ")":
                # if closing decrememtn the count
                leftMin, leftMax = leftMin - 1, leftMax - 1
            else:
                # if wildcard can be considered as left or right parantheisis
                # so we decrement left min considering as closing paranthesis
                # and incremenmt left max considering opening paranthesis
                leftMin, leftMax = leftMin - 1, leftMax + 1
            # max cannot be less than zero as we need opening paranthesis to be more
            if leftMax < 0:
                return False
                # if leftMin goes negative, make it zero as we chose it wrong to have closing parantesis as wildcard
            if leftMin < 0:  # required because -> s = ( * ) (
                leftMin = 0
                # if we have closed all opening paramthesis
        return leftMin == 0



s = "(*))"
checkValidString(s)
```
