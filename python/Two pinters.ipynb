```python
# https://leetcode.com/problems/3sum/
# 15. 3Sum


def threeSum(nums):
  res = []
  nums.sort()
# brute for solution will be using three loops
  for i,a in enumerate(nums):
    if i> 0 and a == nums[i-1]:
      continue
    l,r = i +1, len(nums)-1
    while l< r:
      threeSUm = a + nums[l]+ nums[r]
      if threeSUm > 0:
        r-=1
      elif threeSUm < 0:
        l+=1
      else:
        res.append([a,nums[l], nums[r]])
        l+=1
        while nums[l] == nums[l-1] and l<r:
          l+=1

  return res

# o(n**2) o(n)
#o(n**2) o(1 or n) depends oin sorting algorithm


# nums = [-1,0,1,2,-1,-4]
nums = [0,0,0]
threeSum(nums)
```

```python
  # https://leetcode.com/problems/trapping-rain-water/
  # 42. Trapping Rain Water

  # at any point, get the max of the left and max of the right, than get min of it
  # then minus this min from the height at that point to get the block fowater

  # leftmax is left most height, rightmax is rightmost height
  # starting from left and right pointers, nothing left to leftmost pointer so nop water
# compare the maxleft and maxright, increment the min one eihter left or right pointer.

def trap(height):

    if not height:
      return 0
    l,r = 0, len(height)-1
    leftmax, rightmax = height[l], height[r]
    res=0
    while l<r:
      if leftmax< rightmax:
        l+=1
        leftmax = max(leftmax, height[l])
        res+=leftmax-height[l]
      else:
        r-=1
        rightmax = max(rightmax, height[r])
        res+=rightmax-height[r]
    return res
# o(n) o(1)




height = [4,2,0,3,2,5]
trap(height)
```
