```python
# https://leetcode.cohttps://leetcode.com/problems/reverse-words-in-a-string-iii/description/m/problems/reverse-words-in-a-string/description/
# 151. Reverse Words in a String

def reverseWords(self, s: str) -> str:
        res , i, n = "",0,len(s)
        while i<n:
            while i<n and s[i] ==" ":
                i+=1
            if i>=n:
                break
            j = i+1
            while j<n and s[j] !=" ":
                j+=1
            sub = s[i:j]
            if len(res) == 0:
                res = sub
            else:
                res = sub + " " + res
            i=j+1
        return res

```

```python
# https://leetcode.com/problems/set-matrix-zeroes/
# 73. Set Matrix Zeroes



def setZeroes(matrix):
        # O(1)
        ROWS, COLS = len(matrix), len(matrix[0])
        # this is for position 0,0
        colZero = False

        # determine which rows/cols need to be zero
        for r in range(ROWS):
            for c in range(COLS):
                # set only first row value to zero for corresponding co;lumn
                if matrix[r][c] == 0:
                    matrix[0][c] = 0
                    # if position is not 0,0
                    if r > 0:
                        # put the column to zero
                        matrix[r][0] = 0
                    else:
                        # put 0,0 to true if we need to makje it zero
                        # that means we have zero in first row
                        colZero = True
# go tyhrought rows and columns
# skip the first as we have kept it for checking
        for r in range(1, ROWS):
            for c in range(1, COLS):
                # if top position for that position is zero
                # or left colum position is zero
                if matrix[0][c] == 0 or matrix[r][0] == 0:
                    matrix[r][c] = 0
# is top position is zero
        if matrix[0][0] == 0:
            # make the column zero
            for r in range(ROWS):
                matrix[r][0] = 0
# then make the columns zero
        if colZero:
            for c in range(COLS):
                matrix[0][c] = 0
        return matrix


matrix = [[1,1,1],[1,0,1],[1,1,1]]
setZeroes(matrix)
```

```python
# https://leetcode.com/problems/rotate-image/
# 48. Rotate Image

def rotate(matrix):
        """
        Do not return anything, modify matrix in-place instead.
        """
        l, r = 0, len(matrix) - 1
        while l < r:
            for i in range(r - l):
                top, bottom = l, r

                # save the topleft
                topLeft = matrix[top][l + i]

                # move bottom left into top left
                matrix[top][l + i] = matrix[bottom - i][l]

                # move bottom right into bottom left
                matrix[bottom - i][l] = matrix[bottom][r - i]

                # move top right into bottom right
                matrix[bottom][r - i] = matrix[top + i][r]

                # move top left into top right
                matrix[top + i][r] = topLeft
            r -= 1
            l += 1



matrix = [[1,2,3],[4,5,6],[7,8,9]]

print(rotate(matrix))
```

```python
# https://leetcode.com/problems/spiral-matrix/
# 54. Spiral Matrix

def spiralOrder(matrix):
        res = []
        # NOTICE THAT reight and bottom are out of bound,
        # means we need that for range
        left, right = 0, len(matrix[0])
        top, bottom = 0, len(matrix)

        while left < right and top < bottom:
            # get every i in the top row
            for i in range(left, right):
                res.append(matrix[top][i])
                # increment top here a s we are done with top row
            top += 1
            # get every i in the right col
            for i in range(top, bottom):
                # take the right columns
                res.append(matrix[i][right - 1])
            right -= 1
            # if we have just one column or onbe row
            if not (left < right and top < bottom):
                break
            # get every i in the bottom row
            for i in range(right - 1, left - 1, -1):
                res.append(matrix[bottom - 1][i])
            bottom -= 1
            # get every i in the left col
            for i in range(bottom - 1, top - 1, -1):
                res.append(matrix[i][left])
            left += 1

        return res


matrix = [[1,2,3],[4,5,6],[7,8,9]]
spiralOrder(matrix)
```

```python
# https://leetcode.com/problems/intersection-of-two-linked-lists/
# 160. Intersection of Two Linked Lists


# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        l1, l2 = headA, headB
        while l1!= l2:
            l1 = l1.next if l1 else headB
            l2 = l2.next if l2 else headA
        return l1
```
