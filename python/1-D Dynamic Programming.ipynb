```python
# https://leetcode.com/problems/climbing-stairs/
# 70. Climbing Stairs
def climbStairs(n):
        # if only three elements
        if n <= 3:
            return n
            # two memory caches
        # two is last element, one is second last element
        one, two = 1,1
        # we do n-1 computations
        for i in range(n-1):
            # to temp new one
            temp = one + two
            # new two will be one
            two = one
            # new one will be temp
            one = temp
        return one
# o(n) o (n)


n = 3
climbStairs(n)
```

```python
# https://leetcode.com/problems/min-cost-climbing-stairs/
# 746. Min Cost Climbing Stairs
def minCostClimbingStairs(cost):
        # 10,15,20
        # we do this as we need to reach to that position
        cost.append(0)
        # 10,15,20,0
# third last element we need to calculayte
        for i in range(len(cost) - 3, -1, -1):
            # will be min of cost of one jump and two jumb
            #cost[i] = min(cost[i]+cost[i + 1], cost[i]+ cost[i + 2])
            # can be reduced to
            cost[i] += min(cost[i + 1], cost[i + 2])
# then we return min value of index 0 or 1
        return min(cost[0], cost[1])
#o(n)
cost = [10,15,20]
minCostClimbingStairs(cost)
```

```python
# https://leetcode.com/problems/min-cost-climbing-stairs/description/
# 746. Min Cost Climbing Stairs
def rob(nums):
        # considering first two, we decide to rob one, the mac is one. then we decide to rob to , the max is 2,
        # now for 3, we can rob 3 or not rob three
        # if we rob three, we get 1 +3 =4
        # if we not rob three, we get max of remainder house which is 2
        # so for 3, we get max of 2 and 4, we get 4
        rob1, rob2 = 0, 0
        # g through the nums
        for n in nums:
            # store the new value in temp which is ,max of addition of current and last house or max of adjacent houses
            temp = max(n + rob1, rob2)
            rob1 = rob2
            rob2 = temp
        # return rob 2
        return rob2
# o(n)

nums = [1,2,3,1]
rob(nums)
```

```python
# https://leetcode.com/problems/house-robber-ii/
# 213. House Robber II

def rob(nums):
        # max of three values
        # first value cover s the edge case if nums had onlu one int
        # first skip the first house and seconfd skip the last house
        return max(nums[0], helper(nums[1:]),helper(nums[:-1]))


# helper function similar to house robber 1
def helper(nums):
        rob1, rob2 = 0, 0

        for n in nums:
            newRob = max(rob1 + n, rob2)
            rob1 = rob2
            rob2 = newRob
        return rob2

nums = [2,3,2]
rob(nums)
```

```python
# https://leetcode.com/problems/longest-palindromic-substring/

# 5. Longest Palindromic Substring

def longestPalindrome(s):
        # expand from position to either side to find longest palindromic sting
        res = ""

        resLen = 0

        for i in range(len(s)):
            # £ expanding from the char itself
            # char is a palindrom efor itself
            # or odd length chars
            # odd length
            # l r are at i initially
            l, r = i, i
            while l >= 0 and r < len(s) and s[l] == s[r]:
                # if len is greater, then update the results
                if (r - l + 1) > resLen:
                    res = s[l : r + 1]
                    resLen = r - l + 1
                l -= 1
                r += 1
# considering two adjacent chars
            # even length
            l, r = i, i + 1
            while l >= 0 and r < len(s) and s[l] == s[r]:
                if (r - l + 1) > resLen:
                    res = s[l : r + 1]
                    resLen = r - l + 1
                l -= 1
                r += 1

        return res



s = "babad"
longestPalindrome(s)
```

```python
# https://leetcode.com/problems/coin-change/
# 322. Coin Change



def coinChange(coins, amount):
        # give any max amount
        dp = [amount + 1] * (amount + 1)
        # set dp of zero to zero
        # taht means to calculate 0 we need zero coins
        dp[0] = 0
# calculate count of coins to all the amounts untill the target amount
        for a in range(1, amount + 1):
            # go through each coin denominations
            for c in coins:
                # if difference is positive
                if a - c >= 0:
                    # set dp of that amount equals to i of that count + the dp of the remainder
                    # return the min
                    dp[a] = min(dp[a], 1 + dp[a - c])
                    # if condition to check it is not the default wvalue stored
        return dp[amount] if dp[amount] != amount + 1 else -1
#O(n**2) o(n)


coins = [1,2,5]
amount = 11

coinChange(coins, amount)

```

```python
# https://leetcode.com/problems/maximum-product-subarray/
# 152. Maximum Product Subarray


def maxProduct(nums):
        # O(n)/O(1) : Time/Memory
        # initialize to first num int as to tackle edge case having just one number
        res = nums[0]
        # min and max to 1
        curMin, curMax = 1, 1
# for each nums, we computer current max and current min
        for n in nums:
# we need to store the temp as we are updating the current max in next step
            tmp = curMax * n
            # n because if [-1,8] min max will be -1 and -1 making -8 as max for 8, but we kjnow 8 is tha max which is n
            curMax = max(n * curMax, n * curMin, n)
            curMin = min(tmp, n * curMin, n)
            # result is max of max
            res = max(res, curMax)
            # return res
        return res

# o(n)o(1)

nums = [2,3,-2,4]
maxProduct(nums)
```

```python
# https://leetcode.com/problems/word-break/
# 139. Word Break




def wordBreak(s, wordDict):
# initialize all the positions to false
# bottum up approach
# check if starting from that position do we have result
        dp = [False] * (len(s) + 1)
        # set last to true as we know we have reached end to true
        dp[len(s)] = True

        for i in range(len(s) - 1, -1, -1):
            for w in wordDict:
                # check if the word lenght is greater than length to compare
                # and also check if the length to compare is same as word
                if (i + len(w)) <= len(s) and s[i : i + len(w)] == w:
                    # we have dp of 4 as true, wo for 4 char word, at index o, we will have dp as true
                    # as we have already calculated for dp of the remainig string
                    dp[i] = dp[i + len(w)]
                    # if we found to word break it, we can stop the inner for loop
                if dp[i]:
                    break

        return dp[0]
#(o(m*n*n)) as we are going through s and then word and also comparing each word in s



s = "applepenapple"
wordDict = ["apple","pen"]
wordBreak(s, wordDict)
```

```python
# https://leetcode.com/problems/longest-increasing-subsequence/
# 300. Longest Increasing Subsequence


def lengthOfLIS(nums):
        # strat from the bottom
        # we see that for last pos
        LIS = [1] * len(nums)
# starting from last value#
# bottum up approach
        for i in range(len(nums) - 1, -1, -1):
            # go through all the subsequenses after it
            for j in range(i + 1, len(nums)):
                # only if i is les than j
                if nums[i] < nums[j]:
                    # get the max of int at i or 1+j
                    LIS[i] = max(LIS[i], 1 + LIS[j])
        return max(LIS)


nums = [10,9,2,5,3,7,101,18]
lengthOfLIS(nums)
```

```python
# https://leetcode.com/problems/partition-equal-subset-sum/
# 416. Partition Equal Subset Sum




def canPartition(nums):
        # if divide by two gives remainder, return false
        if sum(nums) % 2:
            return False
# dp set to add the values
        dp = set()
        # add zero first
        dp.add(0)
        # target to reach is half of the sum
        target = sum(nums) // 2
# begin from the end
        for i in range(len(nums) - 1, -1, -1):
            # clreate a new set
            nextDP = set()
            # go through each element in dp
            for t in dp:
                # add the sum to next dp
                nextDP.add(t + nums[i])
                # also add the t as we are ding next steps and can miss t in dp
                nextDP.add(t)
            # make dp as next dp
            dp = nextDP
        return True if target in dp else False



nums = [1,5,11,5]
canPartition(nums)
```
