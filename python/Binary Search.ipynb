```python
# https://leetcode.com/problems/binary-search/submissions/
# 704. Binary Search

def search(nums,target) :
        l,r = 0,len(nums)-1
        while l<=r:
            m = (l+r)//2
            if nums[m] > target:
                r=m-1
            elif nums[m]< target:
                l=m+1
            else:
                return m
        return -1
# o(logn) o(1)

nums = [-1,0,3,5,9,12]
target = 9

search(nums,target)
```

```python
# https://leetcode.com/problems/search-a-2d-matrix/
# 74. Search a 2D Matrix
def searchMatrix(matrix,target):
        if not matrix:
            return False
        m, n = len(matrix), len(matrix[0])
        left, right = 0, m * n - 1

        while left <= right:
            mid = (left + right) // 2
            mid_row, mid_col = divmod(mid, n)

            if matrix[mid_row][mid_col] == target:
                return True
            elif matrix[mid_row][mid_col] < target:
                left = mid + 1
            else:
                right = mid - 1

        return False
# o(log(m*n)) o(1)
matrix = [[1,3,5,7],[10,11,16,20],[23,30,34,60]]
target = 3

searchMatrix(matrix, target)




```

```python
#https://leetcode.com/problems/find-minimum-in-rotated-sorted-array/
#153. Find Minimum in Rotated Sorted Array in o(logn)
# find mid using binary search, compare that with the left ost element. if mid is smaller that left most element, then search right
# else search in left
def findMin(nums):
  # can also be set to infinity
   res = nums[0]
   l,r = 0, len(nums)-1
   while l<=r:
    # if sorted
    if nums[l]< nums[r]:
      res = min(res, nums[l])
      break
    m = (l+r)//2
    res = nums[m]
    # search right if mid is greater than leftmost
    if nums[m]>= nums[l]:
      l = m+1
    else:
      # search left
      r = m-1
   return res
# o(log n) o(1)

nums = [3,4,5,0,1,2]
findMin(nums)
```

```python
# https://leetcode.com/problems/median-of-two-sorted-arrays/
# 4. Median of Two Sorted Arrays

def findMedianSortedArrays(nums1, nums2):
  # create copies
        A, B = nums1, nums2
        # get total length
        total = len(nums1) + len(nums2)
        # get half
        half = total // 2
# we need A of smaller length for algorithm to work
        if len(B) < len(A):
            A, B = B, A

        l, r = 0, len(A) - 1
        while True:
          # get mid of A
            i = (l + r) // 2  # A
            # set partition for b as half minus mid of A
            # minus two to compensate for index of i and j starting from zero
            j = half - i - 2  # B
# if out of bound, set default values
            Aleft = A[i] if i >= 0 else float("-infinity")
            Aright = A[i + 1] if (i + 1) < len(A) else float("infinity")
            Bleft = B[j] if j >= 0 else float("-infinity")
            Bright = B[j + 1] if (j + 1) < len(B) else float("infinity")

            # partition is correct
            if Aleft <= Bright and Bleft <= Aright:
                # odd length of total
                if total % 2:
                  # we will return min of the two
                    return min(Aright, Bright)
                # even
                # get the max of left and min of the rights
                return (max(Aleft, Bleft) + min(Aright, Bright)) / 2
                # if the first condition does not satisfy
                # means we have bigger left partition
            elif Aleft > Bright:
                r = i - 1
                # means we have smaller left partition
            else:
                l = i + 1


nums1 = [1,3]
nums2 = [2]
findMedianSortedArrays(nums1, nums2)
```
