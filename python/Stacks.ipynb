```python
# https://leetcode.com/problems/valid-parentheses/submissions/
# 20. Valid Parentheses

def isValid(s):
        stack = []
        close_open = {
            ')': '(',
            '}': '{',
            ']': '['
        }

        for ch in s:
            if ch in close_open:
                if stack and stack[-1] == close_open[ch]:
                    stack.pop()
                else:
                    return False
            else:
                stack.append(ch)
        return not bool(stack)

      #o(n) o(n)


s = "()"
isValid(s)
```

```python
# https://leetcode.com/problems/min-stack/
# 155. Min Stack

class MinStack:

    def __init__(self):
        self.stack=[]
        self.minStack=[]

    def push(self, val: int) -> None:
        self.stack.append(val)
        val= min(val, self.minStack[-1] if self.minStack else val)
        self.minStack.append(val)

    def pop(self) -> None:
        self.stack.pop()
        self.minStack.pop()

    def top(self) -> int:
        print(self.stack[-1])

    def getMin(self) -> int:
        print(self.minStack[-1])

obj = MinStack()
obj.push(-2)
obj.push(4)
obj.pop()
param_3 = obj.top()
param_4 = obj.getMin()
```
