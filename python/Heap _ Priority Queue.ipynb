```python
# https://leetcode.com/problems/last-stone-weight/
# 1046. Last Stone Weight
import heapq

# simulate mAxheap by pushing negative of the number
def lastStoneWeight(stones):
  # convert to -ve values
  stones = [ -s for s in stones]
  # heapify having the lrgetst value at top as we ive values
  heapq.heapify(stones)
# min twpo stones req
  while len(stones) > 1:
    # pop two heavy stones
    first = heapq.heappop(stones)
    second = heapq.heappop(stones)
    # do math and push to heap
    if second > first:
      heapq.heappush(stones, first-second)
      #edge case stones = [1]
  stones.append(0)

  return abs(stones[0])





stones = [2,7,4,1,8,1]
# stones = [1]
lastStoneWeight(stones)
```

```python
# https://leetcode.com/problems/k-closest-points-to-origin/
# 973. K Closest Points to Origin


def kClosest(points,k):
  minHeap = []
  # go throught the list of cordinates
  for x,y in points:
    # sqrt is not imp as sqrt of larger number will be larger than the sqrt of smaller number
    # and we are doing the minheap to have smaller squares forst
    dist = (x**2)+ (y**2)

    minHeap.append([dist, x, y])
    # get the minHeap using the first value
  heapq.heapify(minHeap)
  res = []
  # get the k min elements and add to the result array
  while k>0:
    dist, x, y = heapq.heappop(minHeap)
    res.append([x, y])
    k-=1

  return res
# o(klogn) o(n)

# points = [[1,3],[-2,2]]
# k = 1
points = [[3,3],[5,-1],[-2,4]]
k = 2
kClosest(points,k)

```

```python
# https://leetcode.com/problems/kth-largest-element-in-an-array/description/
# 215. Kth Largest Element in an Array


def findKthLargest(nums, k):
        k = len(nums) - k
# quickselect algorithm
        def quickSelect(l,r):
            pivot, p = nums[r], l
            for i in range(l,r):
                if nums[i]<= pivot:
                    nums[p], nums[i] = nums[i], nums[p]
                    p+=1
            nums[p], nums[r] = nums[r], nums[p]

            if p>k:
                return quickSelect(l, p-1)
            elif p<k:
                return quickSelect(p+1, r)
            else:
                return nums[p]

        return quickSelect(0, len(nums)-1)


        # o(n) in average case, depends on selection of pivot




      # can also be done by heap in (n-k)logn time

nums = [3,2,1,5,6,4]
k = 2
findKthLargest(nums,k)
```

```python
# https://leetcode.com/problems/task-scheduler/
# 621. Task Scheduler
from collections import Counter, deque
def leastInterval(tasks, n):
        # get the count of each char
        count = Counter(tasks)
        # create max heap with making the numbres -ve
        maxHeap = [-cnt for cnt in count.values()]

        heapq.heapify(maxHeap)

        time = 0
        q = deque()
        # either is empty
        while maxHeap or q:
            # increment time at first
            time+=1
            # check if maxHeap not empty
            if maxHeap:
                # remove the frequent element which will be at top as we have negative numbers
                # pop that and reduce the count by adding 1
                cnt = 1+ heapq.heappop(maxHeap)
                # check if the count is not zero
                if cnt :
                    # append to the queue the count and the time it can be pushed
                    q.append([cnt, time+n])
                # check if we are th the time which is equal to the first element time in the queue
            if q and q[0][1] == time:
                # push the remaining count to the heap again
                heapq.heappush(maxHeap, q.popleft()[0])

            # return the time
        return time



tasks = ["A","A","A","B","B","B"]
n = 2

# tasks = ["A","A","A","B","B","B"]
# n = 0

leastInterval(tasks, n)
```

```python
# https://leetcode.com/problems/design-twitter/description/
# 355. Design Twitter
# MEDIUM

class Twitter:

    def __init__(self):
        # to count the time
        self.count = 0
        # user to tweet map
        self.tweetMap = defaultdict(list)
        # fowwer to followees map
        # set over list is remove is o(1) than o(n)
        self.followMap = defaultdict(set)


    def postTweet(self, userId: int, tweetId: int) -> None:
        # create a tweet, add to tweet map along witht the time it is created
        self.tweetMap[userId].append([self.count, tweetId])
        # minus as we are creating the maxHeap
        self.count-=1



    def getNewsFeed(self, userId: int) -> List[int]:
        res =[]
        minHeap = []
        # add self as the follower of self, this makes it ewasy to get their own tweet using the time
        self.followMap[userId].add(userId)
        # for all the followees for a follower, get their most recent tweets
        for followeeId in self.followMap[userId]:
            if followeeId in self.tweetMap:
                index = len(self.tweetMap[followeeId]) - 1
                count, tweetId = self.tweetMap[followeeId][index]
                # just append themmt ot the minHEap
                # append the time count, tweetid, foloweeid, and the next index to look for most recent tweet
                minHeap.append([count, tweetId, followeeId, index -1])
                # heapify now
        heapq.heapify(minHeap)
        # we need total 10 recent tweets, sontil we have less than 10 tweets and also minHeap
        while minHeap and len(res)<10:
            # pop the max or latest tweet
            count, tweetId, followeeId, index = heapq.heappop(minHeap)
            # append to the result
            res.append(tweetId)
            # check if ther eare any more tweets fop that user
            if index>=0:
                # get the tweet on that index
                count, tweetId = self.tweetMap[followeeId][index]
                # push to heap the required values with idex -1
                heapq.heappush(minHeap,[count, tweetId, followeeId, index -1] )
        return res


    def follow(self, followerId: int, followeeId: int) -> None:
        # add to value set
        # set over list is remove is o(1) than o(n)
        self.followMap[followerId].add(followeeId)


    def unfollow(self, followerId: int, followeeId: int) -> None:
        # if there is a follower list, remove to the list
        if followeeId in self.followMap[followerId]:
            self.followMap[followerId].remove(followeeId)



# Your Twitter object will be instantiated and called as such:
# obj = Twitter()
# obj.postTweet(userId,tweetId)
# param_2 = obj.getNewsFeed(userId)
# obj.follow(followerId,followeeId)
# obj.unfollow(followerId,followeeId)


Example 1:

Input
["Twitter", "postTweet", "getNewsFeed", "follow", "postTweet", "getNewsFeed", "unfollow", "getNewsFeed"]
[[], [1, 5], [1], [1, 2], [2, 6], [1], [1, 2], [1]]
Output
[null, null, [5], null, null, [6, 5], null, [5]]

Explanation
Twitter twitter = new Twitter();
twitter.postTweet(1, 5); // User 1 posts a new tweet (id = 5).
twitter.getNewsFeed(1);  // User 1's news feed should return a list with 1 tweet id -> [5]. return [5]
twitter.follow(1, 2);    // User 1 follows user 2.
twitter.postTweet(2, 6); // User 2 posts a new tweet (id = 6).
twitter.getNewsFeed(1);  // User 1's news feed should return a list with 2 tweet ids -> [6, 5]. Tweet id 6 should precede tweet id 5 because it is posted after tweet id 5.
twitter.unfollow(1, 2);  // User 1 unfollows user 2.
twitter.getNewsFeed(1);  // User 1's news feed should return a list with 1 tweet id -> [5], since user 1 is no longer following user 2.
```

```python
# https://leetcode.com/problems/find-median-from-data-stream/
# 295. Find Median from Data Stream

# HARD


class MedianFinder:
    def __init__(self):
        """
        initialize your data structure here.
        """
        # two heaps, large, small, minheap, maxheap
        # heaps should be equal size
        self.small, self.large = [], []  # maxHeap, minHeap (python default)

    def addNum(self, num: int) -> None:
        # if large and smallest of large is smaller than the num
        # then add to large
        if self.large and num > self.large[0]:
            heapq.heappush(self.large, num)
        else:
            # else add to the small
            heapq.heappush(self.small, -1 * num)
# difference of lengths should atmost be 1
# if small has moere elements, then pop from it and add to the large
        if len(self.small) > len(self.large) + 1:
            val = -1 * heapq.heappop(self.small)
            heapq.heappush(self.large, val)
            # ekse pop from large and add to the small
        if len(self.large) > len(self.small) + 1:
            val = heapq.heappop(self.large)
            heapq.heappush(self.small, -1 * val)

    def findMedian(self) -> float:
        # len f small is larget, then return its smaller element which is large
        if len(self.small) > len(self.large):
            return -1 * self.small[0]
            # else return smaller of large
        elif len(self.large) > len(self.small):
            return self.large[0]
        # else return mid of top of small and large
        return (-1 * self.small[0] + self.large[0]) / 2.0






Example 1:

Input
["MedianFinder", "addNum", "addNum", "findMedian", "addNum", "findMedian"]
[[], [1], [2], [], [3], []]
Output
[null, null, null, 1.5, null, 2.0]

Explanation
MedianFinder medianFinder = new MedianFinder();
medianFinder.addNum(1);    // arr = [1]
medianFinder.addNum(2);    // arr = [1, 2]
medianFinder.findMedian(); // return 1.5 (i.e., (1 + 2) / 2)
medianFinder.addNum(3);    // arr[1, 2, 3]
medianFinder.findMedian(); // return 2.0


```
