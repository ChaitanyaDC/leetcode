```python
# https://leetcode.com/problems/product-of-array-except-self/
# 238. Product of Array Except Self

def productExceptSelf(nums):
  # res = []
  # product = 1
  # for n in nums:
  #   product*=n
  # for m in nums:
  #   res.append(product//m)
  # print(product)
  # return res

  res = [1] * len(nums)
  prefix = 1
  for i in range(len(nums)):
    res[i]= prefix
    prefix*=nums[i]
  postfix=1
  for j in range(len(nums)-1,-1,-1):
    res[j]*=postfix
    postfix*=nums[j]
  return res
  # o(n) o(1)

nums = [1,2,3,4]
productExceptSelf(nums)
```

```python
import collections
# https://leetcode.com/problems/valid-sudoku/
# 36. Valid Sudoku

def isValidSudoku(board):
  row = collections.defaultdict(list)
  col = collections.defaultdict(list)
  sq = collections.defaultdict(list)
  print(row,col,sq)

  for i in range(9):
    for j in range(9):
      if board[i][j]==".":
        continue
      if (board[i][j] in row[i] or board[i][j] in col[j] or board[i][j] in sq[(i//3,j//3)]):
        return False
      row[i].append(board[i][j])
      col[j].append(board[i][j])
      sq[(i//3,j//3)].append(board[i][j])
  return True

# O(9*9)



# board = [["5","3",".",".","7",".",".",".","."]
# ,["6",".",".","1","9","5",".",".","."]
# ,[".","9","8",".",".",".",".","6","."]
# ,["8",".",".",".","6",".",".",".","3"]
# ,["4",".",".","8",".","3",".",".","1"]
# ,["7",".",".",".","2",".",".",".","6"]
# ,[".","6",".",".",".",".","2","8","."]
# ,[".",".",".","4","1","9",".",".","5"]
# ,[".",".",".",".","8",".",".","7","9"]]

board = [["8","3",".",".","7",".",".",".","."]
,["6",".",".","1","9","5",".",".","."]
,[".","9","8",".",".",".",".","6","."]
,["8",".",".",".","6",".",".",".","3"]
,["4",".",".","8",".","3",".",".","1"]
,["7",".",".",".","2",".",".",".","6"]
,[".","6",".",".",".",".","2","8","."]
,[".",".",".","4","1","9",".",".","5"]
,[".",".",".",".","8",".",".","7","9"]]

isValidSudoku(board)
```

```python
# https://leetcode.com/problems/longest-consecutive-sequence/
# 128. Longest Consecutive Sequence

def longestConsecutive(nums):
  # solution one
  # count=1
  # for num in nums:
  #   if (num -1) in nums:
  #     count+=1

  # return count
  # # O(n) o(1)

  numSet = set(nums)
  count = 0
  for n in nums:
    if (n-1) not in numSet:
      length = 0
      while n+length in numSet:
        length+=1
      count = max(count,length)
  return count


nums = [100,4,200,1,3,2]
# nums = [0,3,7,2,5,8,4,6,0,1]

longestConsecutive(nums)
```

```python
# 271. encode decode a string

def encode(strs):
  res = ""
  for i in strs:
    res+= str(len(i)) + "#" + i
  return res

def decode(s):
  i=0
  res=[]
  while i < len(s):
    length=int(s[i])
    if s[i+1] == "#":
      res.append(s[int(i+1+1):int(i+1+length+1)])
      i+=(1+length+1)
  return res


strs =["neet",'co#de']
encode(strs)
s = "4#neet5#co#de"
decode(s)

```
